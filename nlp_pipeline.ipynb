```python
import nltk
```

names are identified using capital letters, so it is important

```python
sentences=""
with open("sample3.txt",encoding='utf-8') as f:
    sentences = f.readlines()
sentences = [x.strip() for x in sentences]
no_of_tokens_before_processing = len(nltk.tokenize.word_tokenize(" ".join(sentences)))

i=0
for sent in sentences:
    if sent.lower() in ['judgement', 'j u d g e m e n t','judgement:', 'judgement :', 'judgment', 'j u d g m e n t','judgment:', 'judgment :', 'order','o r d e r', 'order:', 'order :']:
        break
    i+=1
sentences = sentences[i+1:]

sentences=nltk.tokenize.sent_tokenize(" ".join([x.strip() for x in sentences]))
sentences
```

```python
no_names_sent_tokens=[]
for sentence in sentences:
    tokens=nltk.tokenize.word_tokenize(sentence)
    named_entities = nltk.ne_chunk(nltk.pos_tag(tokens), binary=False)
    names = []
    for tree in named_entities:
        if hasattr(tree, 'label'):
            #print(tree.label(),[i[0] for i in tree.leaves()])
            for leaf in tree.leaves():
                names.append(leaf[0])
    no_names_sent_tokens.append(" ".join([token for token in tokens if token not in names]))
no_names_sent_tokens
```

```python
no_names_sent_tokens_1s = []
for line in no_names_sent_tokens:
    t=nltk.tokenize.word_tokenize(line)
    t=[i for i in t if len(i)>1]
    t=" ".join(t)
    if len(t)>0:
        no_names_sent_tokens_1s.append(t)
no_names_sent_tokens_1s
```

```python
stopwords = set(nltk.corpus.stopwords.words('english'))

no_names_sent_tokens_1s_sw = []
for line in no_names_sent_tokens_1s:
    words = nltk.word_tokenize(line)
    no_names_sent_tokens_1s_sw.append(" ".join([word for word in words if word.lower() not in stopwords]))

no_of_tokens_after_processing=len(nltk.tokenize.word_tokenize(" ".join(no_names_sent_tokens_1s_sw)))


with open("test_text_summarized.txt", "w", encoding='utf-8') as f:
    for line in no_names_sent_tokens_1s_sw:
        f.write(line + "\n")
no_names_sent_tokens_1s_sw
```
